```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(plt.style.available)
# Set style for better visualizations
plt.style.use('classic')
sns.set_palette('husl')
```

## 1. Basic Matplotlib Plotting

### Common Plot Types
- Line plots
- Scatter plots
- Bar plots
- Histograms

```python
# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with multiple subplots
plt.figure(figsize=(15, 5))

# Line plot
plt.subplot(131)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Scatter plot
plt.subplot(132)
plt.scatter(x[::5], y1[::5], label='sin(x)')
plt.scatter(x[::5], y2[::5], label='cos(x)')
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Histogram
plt.subplot(133)
plt.hist(y1, bins=20, alpha=0.5, label='sin(x)')
plt.hist(y2, bins=20, alpha=0.5, label='cos(x)')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

## 2. Advanced Matplotlib Features

```python
# Create more complex visualizations

# Generate sample data
np.random.seed(42)
data = np.random.randn(1000)

# Create figure and axis objects with subplots()
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

# Customized histogram
ax1.hist(data, bins=30, color='skyblue', edgecolor='black')
ax1.set_title('Histogram with Custom Styling')
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency')

# Box plot
ax2.boxplot([data[::2], data[1::2]], labels=['Even', 'Odd'])
ax2.set_title('Box Plot')
ax2.set_ylabel('Value')

# Violin plot
ax3.violinplot([data[::2], data[1::2]])
ax3.set_title('Violin Plot')
ax3.set_xticks([1, 2])
ax3.set_xticklabels(['Even', 'Odd'])

# Filled line plot
x = np.linspace(0, 10, 100)
y = np.sin(x) * np.exp(-x/5)
ax4.fill_between(x, y, alpha=0.5)
ax4.plot(x, y, 'r--')
ax4.set_title('Filled Line Plot')

plt.tight_layout()
plt.show()
```

## 3. Seaborn Visualizations

```python
# Create sample dataset
np.random.seed(42)
df = pd.DataFrame({
    'group': np.repeat(['A', 'B', 'C', 'D'], 25),
    'values': np.random.randn(100),
    'category': np.random.choice(['cat1', 'cat2', 'cat3'], 100),
    'size': np.random.uniform(10, 100, 100)
})

# Create multiple Seaborn plots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# Box plot
sns.boxplot(data=df, x='group', y='values', ax=ax1)
ax1.set_title('Box Plot')

# Violin plot
sns.violinplot(data=df, x='group', y='values', hue='category', ax=ax2)
ax2.set_title('Violin Plot with Categories')

# Scatter plot
sns.scatterplot(data=df, x='values', y='size', hue='group', size='size', ax=ax3)
ax3.set_title('Scatter Plot with Size and Color Encoding')

# KDE plot
sns.kdeplot(data=df, x='values', hue='group', ax=ax4)
ax4.set_title('KDE Plot by Group')

plt.tight_layout()
plt.show()
```

## 4. Statistical Visualizations with Seaborn

```python
# Create a more complex dataset
np.random.seed(42)
df_stats = pd.DataFrame({
    'x': np.random.normal(0, 1, 100),
    'y': np.random.normal(0, 1, 100),
    'category': np.random.choice(['A', 'B', 'C'], 100),
    'value': np.random.uniform(0, 100, 100)
})

# Create statistical visualizations
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# Joint plot
sns.jointplot(data=df_stats, x='x', y='y', hue='category')

# Regression plot
sns.regplot(data=df_stats, x='x', y='value', ax=ax1)
ax1.set_title('Regression Plot')

# Heat map
correlation = df_stats.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', ax=ax2)
ax2.set_title('Correlation Heatmap')

# Pair plot
sns.pairplot(df_stats, hue='category')

plt.tight_layout()
plt.show()
```

## Practical Exercises

```python
# Exercise 1: Sales Data Visualization

# Create sample sales data
np.random.seed(42)
dates = pd.date_range('2023-01-01', '2023-12-31', freq='D')
sales_data = pd.DataFrame({
    'Date': dates,
    'Sales': np.random.normal(1000, 200, len(dates)),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Food'], len(dates)),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], len(dates))
})

def analyze_sales(df):
    # Time series plot
    plt.figure(figsize=(15, 10))

    plt.subplot(221)
    sns.lineplot(data=df, x='Date', y='Sales')
    plt.title('Daily Sales Trend')
    plt.xticks(rotation=45)

    # Sales by category
    plt.subplot(222)
    sns.boxplot(data=df, x='Category', y='Sales')
    plt.title('Sales Distribution by Category')

    # Sales by region
    plt.subplot(223)
    sns.barplot(data=df.groupby('Region')['Sales'].mean().reset_index(),
                x='Region', y='Sales')
    plt.title('Average Sales by Region')

    # Monthly trend
    plt.subplot(224)
    monthly_sales = df.groupby(df['Date'].dt.month)['Sales'].mean()
    sns.lineplot(data=monthly_sales)
    plt.title('Monthly Sales Trend')
    plt.xlabel('Month')

    plt.tight_layout()
    plt.show()

analyze_sales(sales_data)
```

```python
# Exercise 2: Customer Analysis Visualization

# Create customer data
np.random.seed(42)
n_customers = 200

customer_data = pd.DataFrame({
    'Age': np.random.normal(40, 15, n_customers),
    'Income': np.random.normal(50000, 20000, n_customers),
    'SpendingScore': np.random.normal(50, 25, n_customers),
    'Satisfaction': np.random.normal(7, 2, n_customers),
    'LoyaltyYears': np.random.uniform(0, 10, n_customers)
})

def analyze_customers(df):
    # Create customer segments
    df['Segment'] = pd.qcut(df['SpendingScore'], q=3, labels=['Low', 'Medium', 'High'])

    # Customer segmentation visualization
    plt.figure(figsize=(15, 10))

    # Age vs Income scatter plot
    plt.subplot(221)
    sns.scatterplot(data=df, x='Age', y='Income', hue='Segment', size='SpendingScore')
    plt.title('Customer Segmentation')

    # Satisfaction distribution
    plt.subplot(222)
    sns.histplot(data=df, x='Satisfaction', hue='Segment', multiple="stack")
    plt.title('Satisfaction Distribution by Segment')

    # Loyalty years vs Spending
    plt.subplot(223)
    sns.regplot(data=df, x='LoyaltyYears', y='SpendingScore')
    plt.title('Loyalty vs Spending Correlation')

    # Age distribution by segment
    plt.subplot(224)
    sns.violinplot(data=df, x='Segment', y='Age')
    plt.title('Age Distribution by Segment')

    plt.tight_layout()
    plt.show()

analyze_customers(customer_data)
```
